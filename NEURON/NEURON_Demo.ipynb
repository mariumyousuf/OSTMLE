This notebook saves a raster plot for $N=5$ neurons. This raster plot is used as input OST-MLE in the demo notebook saved as "../demo_notebooks/demo_paramRecov_NEURON.ipynb"  

```python
import numpy as np
import matplotlib.pyplot as plt
import sys, os
import subprocess

sys.path.append(os.path.abspath("../scripts/"))
from helpers import *
from osmle import *
```

```python
N=5 # number of neurons
duration_in_s=5 # duration of recording in seconds
numSamples=1 # number of trials (raster plots) to run/save
h_ms = 6 # binarizing window 

## The ground truth was generated as follows, network connectivity matching that of fig2AD ###
w_edge=0.02 # average influence to use in log-normal 
motifs = [
    {"type": "cycle", "nodes": [0,1,2]},
    {"type": "single_edge", "nodes": [2,3]},
    {"type": "empty", "nodes": [4]}
]

# a std of ~0.3–0.5 × mean is biologically plausible
W_gt = generate_W_gt(N, motifs=motifs, w_mean=w_edge, w_std=0.3*w_edge, seed=42)
```

```python
basefolder=f"{N}_toyexample"
gt_fn=f"../demo_gt_for_NEURON/{basefolder}.txt" # path to folder where to retrieve the connectivity matrix from
np.savetxt(gt_fn, W_gt, header=f"{N}\n{N}", comments='')
rasterplot_path=f"../demo_data_from_NEURON/{basefolder}" # path to folder where to save the raster plot generated by NEURON 
```

```python
## run NEURON script
subprocess.run(
    [
        sys.executable,
        "run_neuron.py",
        str(N),
        str(duration_in_s),
        gt_fn,
        rasterplot_path,
        str(numSamples),
    ],
    check=True
)
```
